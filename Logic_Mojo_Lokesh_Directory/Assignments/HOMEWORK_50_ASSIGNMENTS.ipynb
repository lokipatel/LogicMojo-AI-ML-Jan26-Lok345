```python
## Problem 1: Image Dimensions Handler

def image_info(dimensions):
    image_prop = {}
    image_prop['height'] = dimensions[0]
    image_prop['width'] = dimensions[1]
    image_prop['channels'] = dimensions[2]
    image_prop['total_pixels'] = image_prop['height'] * image_prop['width']
    image_prop['is_grayscale'] = True if image_prop['channels'] == 1 else False
    
    return image_prop

print (image_info((224, 224, 3)))
```

```python
## Problem 2: Coordinate Distance Calculator 

import math

def distance(point1, point2):
    x_dist = pow((point1[0]-point2[0]),2)
    y_dist = pow((point1[1]-point2[1]),2)
    
    total_dist = pow(x_dist+y_dist, 0.5)
    return total_dist

# Test your solution
print(distance((0, 0), (3, 4)))
print(distance((1, 2), (4, 6)))
```

```python
## Problem 3: Training Configuration Validator

def validate_config(config):
    learning_rate = config[0]
    batch_size = config[1]
    epochs = config[2]
    optimizer = config[3]
    
    remainder = int(math.log2(batch_size))
    
    error_msg = []
    is_valid = 0
    if not (0.0001 <= learning_rate <= 1.0):
        error_msg.append("learning_rate out of range")
        is_valid = 1
    if not (pow(2, remainder) == batch_size):
        error_msg.append("batch_size not power of 2")
        is_valid = 1
    if not (epochs > 0):
        error_msg.append("epochs must be positive")
        is_valid = 1
    if optimizer not in ['adam', 'sgd', 'rmsprop']:
        error_msg.append("invalid optimizer")
        is_valid = 1
    
    is_valid_bool = True if is_valid == 0 else False
    
    return (is_valid_bool, error_msg)

# Test your solution
print(validate_config((0.001, 32, 100, 'adam')))
print(validate_config((2.0, 30, -5, 'bad_optimizer')))
```
