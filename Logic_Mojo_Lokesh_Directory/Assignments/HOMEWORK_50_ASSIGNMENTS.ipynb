```python
## Problem 1: Image Dimensions Handler

def image_info(dimensions):
    image_prop = {}
    image_prop['height'] = dimensions[0]
    image_prop['width'] = dimensions[1]
    image_prop['channels'] = dimensions[2]
    image_prop['total_pixels'] = image_prop['height'] * image_prop['width']
    image_prop['is_grayscale'] = True if image_prop['channels'] == 1 else False
    
    return image_prop

print (image_info((224, 224, 3)))
```

```python
## Problem 2: Coordinate Distance Calculator 

import math

def distance(point1, point2):
    x_dist = pow((point1[0]-point2[0]),2)
    y_dist = pow((point1[1]-point2[1]),2)
    
    total_dist = pow(x_dist+y_dist, 0.5)
    return total_dist

# Test your solution
print(distance((0, 0), (3, 4)))
print(distance((1, 2), (4, 6)))
```

```python
## Problem 3: Training Configuration Validator

def validate_config(config):
    learning_rate = config[0]
    batch_size = config[1]
    epochs = config[2]
    optimizer = config[3]
    
    remainder = int(math.log2(batch_size))
    
    error_msg = []
    is_valid = 0
    if not (0.0001 <= learning_rate <= 1.0):
        error_msg.append("learning_rate out of range")
        is_valid = 1
    if not (pow(2, remainder) == batch_size):
        error_msg.append("batch_size not power of 2")
        is_valid = 1
    if not (epochs > 0):
        error_msg.append("epochs must be positive")
        is_valid = 1
    if optimizer not in ['adam', 'sgd', 'rmsprop']:
        error_msg.append("invalid optimizer")
        is_valid = 1
    
    is_valid_bool = True if is_valid == 0 else False
    
    return (is_valid_bool, error_msg)

# Test your solution
print(validate_config((0.001, 32, 100, 'adam')))
print(validate_config((2.0, 16, -5, 'bad_optimizer')))
```

```python
## Problem 4: Data Split Generator

def split_data(total_samples, train_ratio, val_ratio, test_ratio):
    # for training
    train_start_index = 0
    train_end_index = int(total_samples * train_ratio)
    
    validation_start_index = train_end_index
    validation_end_index = train_end_index + int(val_ratio * total_samples)
    
    test_start_index = validation_end_index
    test_end_index = validation_end_index + int (test_ratio * total_samples)
    
    return (
        (train_start_index, train_end_index), 
        (validation_start_index, validation_end_index),
        (test_start_index, test_end_index)
                      )

# Test your solution
print(split_data(100, 0.7, 0.2, 0.1))
print(split_data(1000, 0.8, 0.1, 0.1))
```

```python
## Model Metrics Comparison
import statistics

def compare_models(models):
    model_list = []
    accuracy = []
    precision = []
    recall = []
    f1_score = []
    
       
    for param in models:
        model_list.append(param[0])
        accuracy.append(param[1])
        precision.append(param[2])
        recall.append(param[3])
        f1_score.append(param[4])
    
    output = {}
    output['best_models'] = {
        'accuracy' : model_list[accuracy.index(max(accuracy))],
        'precision' : model_list[precision.index(max(precision))],
        'recall' : model_list[recall.index(max(recall))],
        'f1_score' : model_list[f1_score.index(max(f1_score))]
    }
    
    output['averages'] = {
        'accuracy' : round(statistics.mean(accuracy),2),
        'precision' : round(statistics.mean(precision),1),
        'recall' : round(statistics.mean(recall),2)
    }
    
    ranked_by_f1 = []
    ordered_f1 = f1_score.copy()
    ordered_f1.sort(reverse=True)
      
    for element in ordered_f1:
        ranked_by_f1.append(model_list[f1_score.index(element)])
    
    output['ranked_by_f1'] = ranked_by_f1
    return output

# Test your solution
models = [
    ('Model_A', 0.85, 0.82, 0.88, 0.85),
    ('Model_B', 0.92, 0.90, 0.85, 0.87),
    ('Model_C', 0.88, 0.85, 0.92, 0.88)
]

print(compare_models(models))
```

```python
## Problem 6: Duplicate Detector

def find_duplicates(data):
    dup = set()
    set_list = set()
    
    for a in data:
        if a in set_list:
            dup.add(a)
        set_list.add(a)
    output = {}
    output = {
        'has_duplicates' : True if len(dup) > 0 else False,
        'unique_count' : len(set_list),
        'duplicate_count' : len(dup),
        'duplicates' : dup
    }
    return output

# Test your solution
print(find_duplicates([1, 2, 3, 2, 4, 3, 5]))
print(find_duplicates([1, 2, 3, 4, 5]))
```

```python
## Problem 7: Feature Engineering - Categorical Encoder

def encode_categories(data):
    # Write your code here
    category_set = set()
    encoding_map = {}
    encoded_data = []
    encoding = 0
    
    for a in data:
        if a not in category_set:
            encoding_map[a] = encoding
            encoding+=1
        encoded_data.append(encoding_map[a])
        category_set.add(a)
    
    output={}
    output = {
        'encoding_map' : encoding_map,
        'encoding_data' : encoded_data,
        'num_categories' : len(category_set)
    }
    return output

# Test your solution
print(encode_categories(['cat', 'dog', 'cat', 'bird', 'dog', 'cat']))
```

```python
## Problem 8: Dataset Overlap Analyzer

def analyze_overlap(train_ids, test_ids):
    train_id_set, test_id_set = set(train_ids), set(test_ids)
    
    total_integers = train_id_set.union(test_id_set)
    overlap_set = train_id_set.intersection(test_id_set)
    
    has_leakage = True if len(overlap_set) > 0 else False
    
    output = {
        'total_unique' : len(total_integers),
        'overlap' : overlap_set,
        'only_train' : train_id_set - overlap_set,
        'only_test' : test_id_set - overlap_set,
        'has_leakage' : has_leakage
    }
    
    return output

# Test your solution
print(analyze_overlap([1, 2, 3, 4, 5], [4, 5, 6, 7, 8]))
```

```python
## Problem 9: Text Preprocessing - Stopword Remover

def remove_stopwords(text, stopwords):
    # Write your code here
    stopwords_set = set(stopwords)
    text_list = text.split()
    text_set = set(text_list)
    
    filtered_words = text_set - stopwords_set
    
    for word in filtered_words:
        if word in text_list:
            text_list.remove(word)
       
    output = {
        'original_words' : text_set,
        'filtered_words' : filtered_words,
        'removed_words' : len(text_list),
        'unique_filtered' : set(filtered_words)
    }
    
    return output

# Test your solution
stopwords = {'the', 'is', 'a', 'an', 'in', 'on', 'at'}
text = "the cat is on the mat in the house"
print(remove_stopwords(text, stopwords))
```

```python
## Problem 10: Feature Selection - Correlation Filter 

from collections import Counter

# YOUR SOLUTION HERE
def filter_correlated(correlations, threshold):
    # Write your code here   
    high_correlated_pairs_set = {
        set_item for set_item, value in correlations.items() if value >= threshold
    }
    
    all_elements_universal_set = {
        set_item for set_item, value in correlations.items()
    }
    
    freq = Counter()
    
    for a,b in high_correlated_pairs_set:
        freq[a]+=1
        freq[b]+=1
    features_to_remove,all_elements_set = set(),set()
    
    for a,b in all_elements_universal_set:
        if freq[a] > 1:
            features_to_remove.add(a)
        if freq[b] > 1:
            features_to_remove.add(b)
        all_elements_set.add(a)
        all_elements_set.add(b)
    
    features_to_keep = all_elements_set - features_to_remove
    
    output = {
        'high_correlated_pairs_set' : high_correlated_pairs_set,
        'features_to_remove' : features_to_remove,
        'features_to_keep' : features_to_keep
    }
    return output

# Test your solution
correlations = {
    ('f1', 'f2'): 0.95,
    ('f1', 'f3'): 0.5,
    ('f2', 'f4'): 0.92,
    ('f3', 'f4'): 0.3
}
print(filter_correlated(correlations, threshold=0.9))
```

```python
## Problem 12: Feature Statistics Calculator 🟡

# YOUR SOLUTION HERE
import statistics

def calculate_stats(data):
    # Write your code here
    # Hint: Use statistics module (mean, median, stdev) and sorted() for quartiles
    output={}
    for feature, value in data.items():
        output[feature]={
            'mean' : statistics.mean(value),
            'median' : statistics.median(value),
            'std' : round(statistics.pstdev(value),2),
            'min' : min(value),
            'max' : max(value),
            'q1' : statistics.quantiles(value, n=4, method='inclusive')[0],
            'q3' : statistics.quantiles(value,n=4, method='inclusive')[2]
        }
    return output

# Test your solution
data = {
    'age': [25, 30, 35, 40, 45],
    'income': [50000, 60000, 70000, 80000, 90000]
}
print(calculate_stats(data))
```

```python
# Problem 13: Data Aggregator 🟡
from collections import Counter
import statistics

# YOUR SOLUTION HERE
def aggregate_data(data):
    # Write your code here
    output = {}
    freq = Counter()
    category_list = {}
    for item in data:
        value_list=[]
        freq[item['category']]+=1
        if item['category'] not in category_list:
            category_list[item['category']] = []
        category_list[item['category']].append(item['value'])
        output[item['category']]={
            'count' : freq[item['category']],
            'sum' : sum(category_list[item['category']]),
            'avg' : round(statistics.mean(category_list[item['category']]),2),
            'values_list' : category_list[item['category']]
        }
    return output

# Test your solution
data = [
    {'category': 'A', 'value': 10},
    {'category': 'B', 'value': 20},
    {'category': 'A', 'value': 15},
    {'category': 'B', 'value': 25},
    {'category': 'A', 'value': 12}
]
print(aggregate_data(data))
```

```python
# Problem 14: Confusion Matrix Builder 🔴
from collections import Counter
from collections import defaultdict

# YOUR SOLUTION HERE
def build_confusion_matrix(actual, predicted):
    combined = zip(actual, predicted)
    matrix = defaultdict(lambda: defaultdict(int))
    output = {}
    correct_predictions = 0
    for a, p in combined:
        matrix[a][p]+=1
        if a == p:
            correct_predictions+=1
        
    per_class = {}
          
    # converting to normal dictionary
    matrix = {a: dict(preds) for a, preds in matrix.items()}
    accuracy = round(correct_predictions/len(actual), 3)
    
    # getting precision, recall and f1 for each class
    for key in matrix.keys():
        TP = matrix.get(key, {}).get(key, 0)
        FP = sum(matrix.get(other, {}).get(key, 0) for other in matrix.keys() if other != key)
        FN = sum(matrix.get(key, {}).get(other, 0) for other in matrix.keys() if other != key)
        
        
        precision = TP / (TP + FP) if TP + FP else 0
        recall = TP / (TP + FN) if TP + FN else 0
        f1 = (2 * precision * recall / (precision + recall)
              if precision + recall else 0)
        per_class[key] = {
            'precision' : round(precision, 2),
            'recall' : round(recall, 2),
            'f1' : round(f1,2)
        }
    output = {
        'matrix' : matrix,
        'accuracy' : accuracy,
        'per_class' : per_class
    }
    return output

# Test your solution
actual = ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C']
predicted = ['A', 'B', 'B', 'C', 'A', 'A', 'C', 'B']
print(build_confusion_matrix(actual, predicted))
```

```python
# Problem 15: Configuration Merger 🔴
def merged_dict(result_dict, source_dict):
    temp_dict = result_dict.copy()
    
    for key, value in source_dict.items():
        # if it is a dictionary then go for recursive
        if key in temp_dict and isinstance(source_dict[key], dict) and isinstance(value, dict):
            temp_dict[key] = merged_dict(temp_dict, value)
        else:
            temp_dict[key] = value
    
    return temp_dict

# YOUR SOLUTION HERE
def merge_configs(configs):
    # Write your code here
    # Hint: Use recursive approach for nested dicts
    output = {}
    
    for config in configs:
        output = merged_dict(output, config)
    
    return output

# Test your solution
configs = [
    {'model': {'layers': 3, 'dropout': 0.5}, 'lr': 0.01},
    {'model': {'layers': 5}, 'batch_size': 32},
    {'model': {'dropout': 0.3}, 'lr': 0.001}
]
print(merge_configs(configs))
```
