```python
## Problem 2: Coordinate Distance Calculator 

import math

def distance(point1, point2):
    x_dist = pow((point1[0]-point2[0]),2)
    y_dist = pow((point1[1]-point2[1]),2)
    
    total_dist = pow(x_dist+y_dist, 0.5)
    return total_dist

# Test your solution
print(distance((0, 0), (3, 4)))
print(distance((1, 2), (4, 6)))
```

```python
## Problem 3: Training Configuration Validator

def validate_config(config):
    learning_rate = config[0]
    batch_size = config[1]
    epochs = config[2]
    optimizer = config[3]
    
    remainder = int(math.log2(batch_size))
    
    error_msg = []
    is_valid = 0
    if not (0.0001 <= learning_rate <= 1.0):
        error_msg.append("learning_rate out of range")
        is_valid = 1
    if not (pow(2, remainder) == batch_size):
        error_msg.append("batch_size not power of 2")
        is_valid = 1
    if not (epochs > 0):
        error_msg.append("epochs must be positive")
        is_valid = 1
    if optimizer not in ['adam', 'sgd', 'rmsprop']:
        error_msg.append("invalid optimizer")
        is_valid = 1
    
    is_valid_bool = True if is_valid == 0 else False
    
    return (is_valid_bool, error_msg)

# Test your solution
print(validate_config((0.001, 32, 100, 'adam')))
print(validate_config((2.0, 16, -5, 'bad_optimizer')))
```

```python
## Problem 4: Data Split Generator

def split_data(total_samples, train_ratio, val_ratio, test_ratio):
    # for training
    train_start_index = 0
    train_end_index = int(total_samples * train_ratio)
    
    validation_start_index = train_end_index
    validation_end_index = train_end_index + int(val_ratio * total_samples)
    
    test_start_index = validation_end_index
    test_end_index = validation_end_index + int (test_ratio * total_samples)
    
    return (
        (train_start_index, train_end_index), 
        (validation_start_index, validation_end_index),
        (test_start_index, test_end_index)
                      )

# Test your solution
print(split_data(100, 0.7, 0.2, 0.1))
print(split_data(1000, 0.8, 0.1, 0.1))
```

```python
## Model Metrics Comparison
import statistics

def compare_models(models):
    model_list = []
    accuracy = []
    precision = []
    recall = []
    f1_score = []
    
       
    for param in models:
        model_list.append(param[0])
        accuracy.append(param[1])
        precision.append(param[2])
        recall.append(param[3])
        f1_score.append(param[4])
    
    output = {}
    output['best_models'] = {
        'accuracy' : model_list[accuracy.index(max(accuracy))],
        'precision' : model_list[precision.index(max(precision))],
        'recall' : model_list[recall.index(max(recall))],
        'f1_score' : model_list[f1_score.index(max(f1_score))]
    }
    
    output['averages'] = {
        'accuracy' : round(statistics.mean(accuracy),2),
        'precision' : round(statistics.mean(precision),1),
        'recall' : round(statistics.mean(recall),2)
    }
    
    ranked_by_f1 = []
    ordered_f1 = f1_score.copy()
    ordered_f1.sort(reverse=True)
      
    for element in ordered_f1:
        ranked_by_f1.append(model_list[f1_score.index(element)])
    
    output['ranked_by_f1'] = ranked_by_f1
    return output

# Test your solution
models = [
    ('Model_A', 0.85, 0.82, 0.88, 0.85),
    ('Model_B', 0.92, 0.90, 0.85, 0.87),
    ('Model_C', 0.88, 0.85, 0.92, 0.88)
]

print(compare_models(models))
```

```python
## Problem 8: Dataset Overlap Analyzer

def analyze_overlap(train_ids, test_ids):
    train_id_set, test_id_set = set(train_ids), set(test_ids)
    
    total_integers = train_id_set.union(test_id_set)
    overlap_set = train_id_set.intersection(test_id_set)
    
    has_leakage = True if len(overlap_set) > 0 else False
    
    output = {
        'total_unique' : len(total_integers),
        'overlap' : overlap_set,
        'only_train' : train_id_set - overlap_set,
        'only_test' : test_id_set - overlap_set,
        'has_leakage' : has_leakage
    }
    
    return output

# Test your solution
print(analyze_overlap([1, 2, 3, 4, 5], [4, 5, 6, 7, 8]))
```

```python
## Problem 9: Text Preprocessing - Stopword Remover

def remove_stopwords(text, stopwords):
    # Write your code here
    stopwords_set = set(stopwords)
    text_list = text.split()
    text_set = set(text_list)
    
    filtered_words = text_set - stopwords_set
    
    for word in filtered_words:
        if word in text_list:
            text_list.remove(word)
       
    output = {
        'original_words' : text_set,
        'filtered_words' : filtered_words,
        'removed_words' : len(text_list),
        'unique_filtered' : set(filtered_words)
    }
    
    return output

# Test your solution
stopwords = {'the', 'is', 'a', 'an', 'in', 'on', 'at'}
text = "the cat is on the mat in the house"
print(remove_stopwords(text, stopwords))
```

```python
## Problem 10: Feature Selection - Correlation Filter 

from collections import Counter

# YOUR SOLUTION HERE
def filter_correlated(correlations, threshold):
    # Write your code here   
    high_correlated_pairs_set = {
        set_item for set_item, value in correlations.items() if value >= threshold
    }
    
    all_elements_universal_set = {
        set_item for set_item, value in correlations.items()
    }
    
    freq = Counter()
    
    for a,b in high_correlated_pairs_set:
        freq[a]+=1
        freq[b]+=1
    features_to_remove,all_elements_set = set(),set()
    
    for a,b in all_elements_universal_set:
        if freq[a] > 1:
            features_to_remove.add(a)
        if freq[b] > 1:
            features_to_remove.add(b)
        all_elements_set.add(a)
        all_elements_set.add(b)
    
    features_to_keep = all_elements_set - features_to_remove
    
    output = {
        'high_correlated_pairs_set' : high_correlated_pairs_set,
        'features_to_remove' : features_to_remove,
        'features_to_keep' : features_to_keep
    }
    return output

# Test your solution
correlations = {
    ('f1', 'f2'): 0.95,
    ('f1', 'f3'): 0.5,
    ('f2', 'f4'): 0.92,
    ('f3', 'f4'): 0.3
}
print(filter_correlated(correlations, threshold=0.9))
```

```python
# Problem 14: Confusion Matrix Builder 🔴
from collections import Counter
from collections import defaultdict

# YOUR SOLUTION HERE
def build_confusion_matrix(actual, predicted):
    combined = zip(actual, predicted)
    matrix = defaultdict(lambda: defaultdict(int))
    output = {}
    correct_predictions = 0
    for a, p in combined:
        matrix[a][p]+=1
        if a == p:
            correct_predictions+=1
        
    per_class = {}
          
    # converting to normal dictionary
    matrix = {a: dict(preds) for a, preds in matrix.items()}
    accuracy = round(correct_predictions/len(actual), 3)
    
    # getting precision, recall and f1 for each class
    for key in matrix.keys():
        TP = matrix.get(key, {}).get(key, 0)
        FP = sum(matrix.get(other, {}).get(key, 0) for other in matrix.keys() if other != key)
        FN = sum(matrix.get(key, {}).get(other, 0) for other in matrix.keys() if other != key)
        
        
        precision = TP / (TP + FP) if TP + FP else 0
        recall = TP / (TP + FN) if TP + FN else 0
        f1 = (2 * precision * recall / (precision + recall)
              if precision + recall else 0)
        per_class[key] = {
            'precision' : round(precision, 2),
            'recall' : round(recall, 2),
            'f1' : round(f1,2)
        }
    output = {
        'matrix' : matrix,
        'accuracy' : accuracy,
        'per_class' : per_class
    }
    return output

# Test your solution
actual = ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C']
predicted = ['A', 'B', 'B', 'C', 'A', 'A', 'C', 'B']
print(build_confusion_matrix(actual, predicted))
```

```python
# Problem 15: Configuration Merger 🔴
def merged_dict(result_dict, source_dict):
    temp_dict = result_dict.copy()
    
    for key, value in source_dict.items():
        # if it is a dictionary then go for recursive
        if key in temp_dict and isinstance(source_dict[key], dict) and isinstance(value, dict):
            temp_dict[key] = merged_dict(temp_dict, value)
        else:
            temp_dict[key] = value
    
    return temp_dict

# YOUR SOLUTION HERE
def merge_configs(configs):
    # Write your code here
    # Hint: Use recursive approach for nested dicts
    output = {}
    
    for config in configs:
        output = merged_dict(output, config)
    
    return output

# Test your solution
configs = [
    {'model': {'layers': 3, 'dropout': 0.5}, 'lr': 0.01},
    {'model': {'layers': 5}, 'batch_size': 32},
    {'model': {'dropout': 0.3}, 'lr': 0.001}
]
print(merge_configs(configs))
```

```python
## Problem 16: Decorator - Execution Timer 🟡

import time

def timer(func):
    def find_execution_time(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        execution_time = (end_time - start_time) * 1000
        print (f"{func.__name__} executed in {round(execution_time, 2)} ms")
        return result
    return find_execution_time

# Test your decorator
@timer
def slow_function(n):
    time.sleep(n)
    return n * 2

result = slow_function(0.5)
print(f"Result: {result}")
```

```python
## Problem 17: Memoization Cache 🔴

# YOUR SOLUTION HERE
def memoize(func):
    hits = 0
    misses = 0
    cache = {}
    
    def wrapper(*args, **kwargs):
        nonlocal hits, misses
            
        if args in cache:
            hits+=1
            result = cache[args]
            return result
        else:
            misses+=1
            result = func(*args)
            cache[args] = result
            return result
        
    def cache_info():
        return {
            'hits' : hits,
            'misses' : misses,
            'size': len(cache)
        } 
        
    wrapper.cache_info = cache_info
    return wrapper

# Test your decorator
@memoize
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print("First call:", fibonacci(10))
print("Second call:", fibonacci(10))
print("Cache info:", fibonacci.cache_info())
```

```python
## Problem 18: Partial Function Application 🟡

# YOUR SOLUTION HERE
def create_metric(metric_name):
    # Write your function factory here
    if metric_name == 'accuracy':
        def accuracy_func(true_val, pred_val):
            correct = 0
            miss = 0
            for a,b in zip(true_val, pred_val):
                if a == b:
                    correct+=1
                else:
                    miss+=1
            return correct/(correct+miss)
        return accuracy_func
    elif metric_name == 'precision':
        def precision_func(true_val, pred_val):
            tp = 0
            fp = 0
            for a,b in zip(true_val, pred_val):
                if a == 1 and b == 1:
                    tp+=1
                elif a == 0 and b == 1:
                    fp+=1
            return tp/(tp+fp)
        return precision_func

# Test your solution
accuracy_fn = create_metric('accuracy')
precision_fn = create_metric('precision')

y_true = [1, 0, 1, 1, 0]
y_pred = [1, 0, 1, 0, 0]

print("Accuracy:", accuracy_fn(y_true, y_pred))
print("Precision:", precision_fn(y_true, y_pred))
```

```python
## Problem 20: Currying 🔴
import inspect

# YOUR SOLUTION HERE
def curry(func):
    # Write your currying function here
    # Hint: Use inspect.signature to get function arity
    def inner_func(*args):
        total_arguments = len(inspect.signature(func).parameters)
        
        if len(args) >= total_arguments:
            return func(*args)
        else:
            return lambda x : inner_func(*args, x)
    return inner_func

# Test your solution
def add_three(a, b, c):
    return a + b + c

curried_add = curry(add_three)
result1 = curried_add(1)(2)(3)
print(f"Curried: {result1}")

add_1 = curried_add(1)
add_1_2 = add_1(2)
result2 = add_1_2(3)
print(f"Partial application: {result2}")
```

```python
## Problem 22: Feature Scaler 🟡

import statistics

def scale_features(data, method):
    # Write your scaling function here
    if method == "minmax":
        max_val = max(data)
        min_val = min(data)
        result = list(map(lambda x : (x-min_val)/(max_val-min_val), data))
    elif method == 'zscore':
        mean = statistics.mean(data)
        std = statistics.pstdev(data)
        result = list(map(lambda x : round((x-mean)/std,2), data))
    elif method == 'robust':
        median = statistics.median(data)
        q_points = statistics.quantiles(data)
        q1 = q_points[0]
        q3 = q_points[2]
        iqr = q3 - q1
        result = list(map(lambda x : round((x-median)/iqr,2), data))
    else:
        result = 'Unknown type'
    return result

# Test your solution
data = [1, 2, 3, 4, 5]
print("Min-Max:", scale_features(data, 'minmax'))
print("Z-score:", scale_features(data, 'zscore'))
print("Robust:", scale_features(data, 'robust'))
```

```python
# Problem 23: reduce() - Custom Aggregations 🟡

# YOUR SOLUTION HERE
from functools import reduce

def product(numbers):
    result = reduce(lambda x,y : x*y, numbers)
    return result

def maximum(numbers):
    # Write using reduce
    result = reduce(lambda x,y: max(x,y), numbers)
    return result

def join_strings(strings, separator):
    # Write using reduce
    result = reduce(lambda x,y: x + separator + y, strings)
    return result

def flatten(nested_list):
    # Write using reduce
    result = reduce(lambda x,y: x+y, nested_list, [])
    return result

# Test your solutions
print("Product:", product([1, 2, 3, 4, 5]))
print("Maximum:", maximum([3, 1, 4, 1, 5, 9, 2, 6]))
print("Join:", join_strings(['Hello', 'World', 'ML'], ' '))
print("Flatten:", flatten([[1, 2], [3, 4], [5, 6]]))
```

```python
# Problem 24: Comprehension Olympics 🔴
from functools import reduce

def transpose(matrix):
    # Write using list comprehension
    # row
    result = [[row[j] for row in matrix] for j in range(len(matrix[0]))]
    return result

def flatten_even(matrix):
    # Write using list comprehension
    result = [row[j] for row in matrix for j in range(len(matrix[0])) if row[j]%2 == 0]
    return result

def cartesian(list1, list2, threshold=5):
    # Write using list comprehension
    result = [(list1[i],list2[j]) for i in range(len(list1)) for j in range(len(list2)) if list1[i] + list2[j] > threshold]
    return result

def invert_dict(d):
    result = {value : [k for k in d if d[k] == value] for value in set(d.values())}
    return result

# Test your solutions
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("Transpose:", transpose(matrix))
print("Flatten even:", flatten_even(matrix))
print("Cartesian:", cartesian([1, 2, 3], [4, 5, 6]))
print("Invert:", invert_dict({'a': 1, 'b': 2, 'c': 1}))
```
