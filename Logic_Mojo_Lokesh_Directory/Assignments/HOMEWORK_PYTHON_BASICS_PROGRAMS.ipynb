## Program 1: Swap Two Numbers

```python
a, b = map(int, input("Enter two numbers separated by a space: ").split())

print (f"The original a is {a}")
print (f"The original b is {b}")

b, a = a, b
# now they are swapped
print ("After Swapping...")
print (f"The current a is {a}")
print (f"The current b is {b}")
```

## Program 14: Prime Check

```python
import math
def is_prime(n):
    if n<= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n%i == 0:
            return False
    return True

n = int(input("Enter the integer: "))
message = "prime" if is_prime(n) else "not prime"
print (f"The number is {message}")
```

## Program 17: Reverse a String

```python
s = input("Enter the string: ")

reverse_string = []

for i in range(len(s)):
    reverse_string.append(s[len(s)-i-1])

print("The reverse string is " + "".join(map(str, reverse_string)))
```

## Program 18: Palindrome String

```python
def check_palindrome(s):
    for i in range(len(s)):
        if (s[i] != s[len(s)-i-1]):
            return False
    return True

s = input("Enter the string: ")
message = "Palindrome" if check_palindrome(s) else "Not Palindrome"
print (message)
```

## Program 19: Count Vowels

```python
vowel = ['a','e', 'i', 'o', 'u']

s = input("Enter the string: ").lower()

total = 0
for i in range(len(s)):
    if (s[i] in vowel):
        total += 1
        
print (f"The total vowels in the string is {total}")
```

## Program 27: Remove Duplicates (Preserve Order)

```python
a = int(input("Enter the number of integers: "))
numbers = list(map(int, input("Enter space-separated integers: ").split()))

numbers = numbers[:a]

seen = []
unique_array = []

for x in numbers:
    if x not in seen:
        seen.append(x)
        unique_array.append(x)

print (unique_array)
```

## Program 28: Second Largest (Distinct)

```python
a = int(input("Enter the number of integers: "))
numbers = list(map(int, input("Enter space-separated integers: ").split()))

numbers = numbers[:a]

distinct_num_array = list(set(numbers))

if len(distinct_num_array) < 2:
    print ("Not Possible")
else:
    distinct_num_array.sort()
    print (f"The second largest distict integer in the array is {distinct_num_array[-2]}")
```
