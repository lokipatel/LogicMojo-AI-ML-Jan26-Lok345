## Program 1: Swap Two Numbers

```python
a, b = map(int, input("Enter two numbers separated by a space: ").split())

print (f"The original a is {a}")
print (f"The original b is {b}")

b, a = a, b
# now they are swapped
print ("After Swapping...")
print (f"The current a is {a}")
print (f"The current b is {b}")
```

## Program 2: Area of a Circle

```python
radius = float(input("Enter the radius length of the circle: "))
pi = 3.14159
area = pi * radius ** 2
print (f"The area of the circle is {area}")
```

## Program 3: Celsius to Fahrenheit

```python
c = float(input("Enter the celsius temperature: "))
f = (c * 9/5) + 32
print (f"The temperature in Fahrenheit is {f}")
```

## Program 4: Simple Interest

```python
P, R, T = map(int, input("Enter three numbers separated by a space: ").split())
SI = (P * R * T) / 100
print (f"The simple interest is {SI}")
```

## Program 5: Even or Odd

```python
n = int(input("Enter the number: "))

if (n%2 == 0):
    print ("The number is even")
else:
    print ("The number is odd")
```

## Program 6: Positive / Negative / Zero

```python
n = int(input("Enter the number: "))

if (n > 0):
    print ("The number is positive")
elif (n < 0):
    print ("The number is negative")
else:
    print ("The number is Zero")
```

## Program 7: Largest of Three Numbers

```python
a, b, c = map(int, input("Enter three numbers separated by a space: ").split())
print (f"The largest number is {max(a,b,c)}")
```

## Program 8: Leap Year Check

```python
n = int(input("Enter the year: "))
a = n % 400
b = n % 4
c = n % 100

if (a == 0):
    print ("its a leap year")
elif (b == 0 and c != 0):
    print ("its a leap year")
else:
    print ("its not a leap year")
```

## Program 9: Grade from Marks

```python
n = int(input("Enter the marks: "))

grade = ''

if (n<=100 and n>90):
    grade = 'A'
elif (n<=89 and n>80):
    grade = 'B'
elif (n<=79 and n>70):
    grade = 'C'
elif (n<= 69 and n>60):
    grade = 'D'
else:
    grade = 'F'

print (f"The grade is {grade}")
```

## Program 10: Multiplication Table

```python
n = int(input("Enter the integer: "))

print (f"The multiplication table for {n} is")
for i in range(1, 6):
    print (f"{n} * {i} = {n}")
```
