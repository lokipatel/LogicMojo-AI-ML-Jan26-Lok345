## Program 1: Swap Two Numbers

```python
a, b = map(int, input("Enter two numbers separated by a space: ").split())

print (f"The original a is {a}")
print (f"The original b is {b}")

b, a = a, b
# now they are swapped
print ("After Swapping...")
print (f"The current a is {a}")
print (f"The current b is {b}")
```

## Program 2: Area of a Circle

```python
radius = float(input("Enter the radius length of the circle: "))
pi = 3.14159
area = pi * radius ** 2
print (f"The area of the circle is {area}")
```

## Program 3: Celsius to Fahrenheit

```python
c = float(input("Enter the celsius temperature: "))
f = (c * 9/5) + 32
print (f"The temperature in Fahrenheit is {f}")
```

## Program 4: Simple Interest

```python
P, R, T = map(int, input("Enter three numbers separated by a space: ").split())
SI = (P * R * T) / 100
print (f"The simple interest is {SI}")
```

## Program 8: Leap Year Check

```python
n = int(input("Enter the year: "))
a = n % 400
b = n % 4
c = n % 100

if (a == 0):
    print ("its a leap year")
elif (b == 0 and c != 0):
    print ("its a leap year")
else:
    print ("its not a leap year")
```

## Program 9: Grade from Marks

```python
n = int(input("Enter the marks: "))

grade = ''

if (n<=100 and n>90):
    grade = 'A'
elif (n<=89 and n>80):
    grade = 'B'
elif (n<=79 and n>70):
    grade = 'C'
elif (n<= 69 and n>60):
    grade = 'D'
else:
    grade = 'F'

print (f"The grade is {grade}")
```

## Program 10: Multiplication Table

## Program 11: Sum of First N Natural Numbers

```python
n = int(input("Enter the integer: "))
s = 0

for i in range(1, n+1):
    s += i
    
print (f"The total sum of first {n} natural numbers is {s}")
```

## Program 12: Factorial of N

```python
while (n := int(input("Enter a greater than 0: "))) <= 0: pass

s = 1
for i in range(1, n+1):
    s = s * i
    
print (f"Factorial of {n} is {s}")
```

## Program 13: Fibonacci (N Terms)

```python
while (n := int(input("Enter a greater than 0: "))) <= 0: pass

f=[0,1]

for i in range(2,n):
    f.append(f[i-2] + f[i-1])
    
print(" ".join(map(str, f)))
```

## Program 14: Prime Check

```python
import math
def is_prime(n):
    if n<= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n%i == 0:
            return False
    return True

n = int(input("Enter the integer: "))
message = "prime" if is_prime(n) else "not prime"
print (f"The number is {message}")
```

## Program 15: Count Digits

```python
f = int(input("Enter the number: "))

s = 0

while f>0:
    s = s + 1
    f = int(f/10)

print (f"The number of digits is {s}")
```
